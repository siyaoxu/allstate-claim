```python
def lossRestore(logloss, ymean, ystd):
    loss = np.exp(logloss*ystd+ymean)
    return loss
```

# 2nd Level XGBoost

```python
def logregobj(preds, dtrain):
    labels = dtrain.get_label()
    con = 2
    x =preds-labels
    grad =con*x / (np.abs(x)+con)
    hess =con**2 / (np.abs(x)+con)**2
    return grad, hess 

def evalerror(preds, dtrain):
    labels = dtrain.get_label()
    return 'mae', mean_absolute_error(np.exp(preds), np.exp(labels))
```

```python
import xgboost as xgb
```

```python
d_train_xgb = xgb.DMatrix(x_train_2nd,label=y_train)
d_val_xgb = xgb.DMatrix(x_val_2nd,label = y_val)
x_val_xgb = xgb.DMatrix(x_val_2nd)
d_test_xgb = xgb.DMatrix(x_test_2nd)
```

```python
num_round = 250

params = {'eval_metric':'mae','max_depth':6,'colsample_bytree':1,'alpha':1,'gamma':0, \
          'eta':0.02,'min_child_weight':0,'subsample':0.8}# 1111, customized MAE obj function, log, rescale, 3000 round
# params = {'eval_metric':'mae','max_depth':4,'colsample_bytree':0.8,'alpha':1,'gamma':0, \
#           'eta':0.02,'min_child_weight':0,'subsample':0.8}# 1104, customized MAE obj function, log, rescale, 3000 round

watchlist  = [(d_val_xgb,'eval'), (d_train_xgb,'train')]
gbt = xgb.train(params, d_train_xgb,num_round,watchlist, obj=logregobj, early_stopping_rounds=300)
# gbt = xgb.train(params, d_train_xgb, num_round, watchlist, early_stopping_rounds=300)
```

```python
xgb_pred = gbt.predict(x_val_xgb)
print(mean_absolute_error(lossRestore(y_val,ymean,ystd),lossRestore(xgb_pred,ymean,ystd)))
```

```python
xlim = [0,23000]
ylim = [0,23000]
plt.figure(figsize=(15,4))
plt.subplot(1,1,1)
plt.scatter(lossRestore(y_val,ymean,ystd),lossRestore(xgb_pred,ymean,ystd));
plt.plot([0,23000],[0,23000],'r');
plt.ylim(ylim)
plt.xlim(xlim)
```
